```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart-disease.csv')
data
```

```python
x, y = data.drop('target', axis=1), np.ravel(data['target'])
```

### 1)

#### a.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold, cross_val_score

predictorKNN = KNeighborsClassifier(n_neighbors = 5)
predictorGaussian = GaussianNB()

cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 0)

predictorKNN_scores = cross_val_score(predictorKNN, x, y, cv = cv, scoring = 'accuracy')
predictorGaussian_scores = cross_val_score(predictorGaussian, x, y, cv = cv, scoring = 'accuracy')

results_df = pd.DataFrame({
    'kNN': predictorKNN_scores,
    'GaussianNB': predictorGaussian_scores
})

plt.figure(figsize = (8,6))
sns.boxplot(data = results_df)
plt.title('Fold Accuracies for kNN (k=5) and Gaussian Naive Bayes')
plt.ylabel('Accuracy')
plt.show()

print("kNN Accuracy:", predictorKNN_scores)
print("GaussianNB Accuracy:", predictorGaussian_scores)

print("\nkNN Mean Accuracy:", predictorKNN_scores.mean())
print("kNN Std Accuracy:", predictorKNN_scores.std())

print("\nGaussianNB Mean Accuracy:", predictorGaussian_scores.mean())
print("GaussianNB Std Accuracy:", predictorGaussian_scores.std())
```

Como vemos pelo grÃ¡fico e pelo desvio padrÃ£o obtidos, o GaussianNB Ã© o mais estÃ¡vel relativamente a performance. Esta observaÃ§Ã£o estÃ¡ de acordo com o esperado, porque o GaussianNB Ã© um modelo probabilÃ­stico (paramÃ©trico),  que assume independÃªncia entre as variÃ¡veis, ajudando na generalizaÃ§Ã£o do modelo. O kNN Ã© baseado na amostra obtida (nÃ£o paramÃ©trico), sujeito a maior variabilidade, estando mais dependente dos dados de treino (pode dar overfit a cada fold).

#### b.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler().fit(x)
x_scaled = scaler.transform(x)

predictorKNN_scores_scaled = cross_val_score(predictorKNN, x_scaled, y, cv = cv, scoring = 'accuracy')
predictorGaussian_scores_scaled = cross_val_score(predictorGaussian, x_scaled, y, cv = cv, scoring = 'accuracy')

results_df = pd.DataFrame({
    'kNN': predictorKNN_scores_scaled,
    'GaussianNB': predictorGaussian_scores_scaled
})

plt.figure(figsize = (8,6))
sns.boxplot(data = results_df)
plt.title('Fold Accuracies (when x is Scaled) for kNN (k=5) and Gaussian Naive Bayes')
plt.ylabel('Accuracy')
plt.show()

print("kNN Accuracy (Min-Max Scaled):", predictorKNN_scores_scaled)
print("GaussianNB Accuracy (Min-Max Scaled):", predictorGaussian_scores_scaled)

print("\nkNN Mean Accuracy (Min-Max Scaled):", predictorKNN_scores_scaled.mean())
print("kNN Std Accuracy (Min-Max Scaled):", predictorKNN_scores_scaled.std())

print("\nGaussianNB Mean Accuracy (Min-Max Scaled):", predictorGaussian_scores_scaled.mean())
print("GaussianNB Std Accuracy (Min-Max Scaled):", predictorGaussian_scores_scaled.std())
```

Como vemos pelo grÃ¡fico e resultados em cima apresentados, o modelo kNN melhorou bastante ao fazer o prÃ©-processamento. Esta observaÃ§Ã£o estÃ¡ de acordo com o esperado, uma vez que o kNN Ã© um modelo baseado em distÃ¢ncias (entre duas instÃ¢ncias). Como cada variÃ¡vel Ã  partida tem uma escala diferente, variÃ¡veis com escalas maiores tÃªm maior impacto na distÃ¢ncia. Ao fazer o Min-Max scaling todas as variÃ¡veis tÃªm a mesma escala 0-1, tendo assim o mesmo impacto na distÃ¢ncia.
Os resultados do modelo GaussianNB permaneceram iguais, porque este modelo nÃ£o Ã© baseado em distÃ¢ncias, assim Ã© irrelevante a escala de cada variÃ¡vel


#### c.

```python
from scipy import stats

res = stats.ttest_rel(predictorKNN_scores_scaled, predictorGaussian_scores_scaled, alternative='less')
print("kNN < GaussianNB? pval =", res.pvalue)
```

NÃ£o conseguimos concluir que o modelo kNN Ã© estatÃ­sticamente superior ao modelo GaussianNB, de acordo com a accuracy, uma vez que o pval > 0.05.
TerÃ­amos de aumentar o intervalo de confianÃ§a para 0.26 para rejeitar a hipÃ³tese nula (kNN < GaussianNB), ou seja, afirmar que o modelo kNN Ã© estatÃ­sticamente superior ao modelo GaussianNB.


### 2)

#### a.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 0)

k_values = [1, 5, 10, 20, 30]
weights_options = ['uniform', 'distance']

train_accuracies = {weight: [] for weight in weights_options}
test_accuracies = {weight: [] for weight in weights_options}

for k in k_values:
    for weight in weights_options:
        knn = KNeighborsClassifier(n_neighbors = k, weights = weight)

        cv_train_scores = cross_val_score(knn, x_train, y_train, cv = cv, scoring = 'accuracy')
        train_accuracies[weight].append(np.mean(cv_train_scores))
        
        knn.fit(x_train, y_train)
        test_accuracy = knn.score(x_test, y_test)
        test_accuracies[weight].append(test_accuracy)

results_df = pd.DataFrame({
    'k': k_values,
    'Train Accuracy (Uniform)': train_accuracies['uniform'],
    'Test Accuracy (Uniform)': test_accuracies['uniform'],
    'Train Accuracy (Distance)': train_accuracies['distance'],
    'Test Accuracy (Distance)': test_accuracies['distance'],
})

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(results_df['k'], results_df['Train Accuracy (Uniform)'], marker='o', label='Train (Uniform)')
plt.plot(results_df['k'], results_df['Test Accuracy (Uniform)'], marker='o', label='Test (Uniform)')
plt.title('kNN Classifier Accuracy (Uniform Weights)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.xticks(results_df['k'])
plt.legend()
plt.ylim(0.50, 0.75)

plt.subplot(1, 2, 2)
plt.plot(results_df['k'], results_df['Train Accuracy (Distance)'], marker='o', label='Train (Distance)')
plt.plot(results_df['k'], results_df['Test Accuracy (Distance)'], marker='o', label='Test (Distance)')
plt.title('kNN Classifier Accuracy (Distance Weights)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.xticks(results_df['k'])
plt.legend()
plt.ylim(0.50, 0.75)

plt.tight_layout()
plt.show()
```

#### b.

Observando o grÃ¡fico (dados nÃ£o escalados), vemos que aumentar o nÃºmero de vizinhos leva a um aumento da accuracy do classificador kNN do grupo de teste, para uniform weights e distance weights. Isto era de esperar, porque nÃ£o nos estamos a limitar a analisar a vizinhanÃ§a prÃ³xima, mas sim a um maior nÃºmero de vizinhos, o que melhora o poder de generalizaÃ§Ã£o do modelo.

```python
plt.figure(figsize=(12, 9))
sns.heatmap(x.corr(method='pearson'), annot=True)
plt.show()
```

Um dos problemas do GaussianNB, para este dataset, Ã© assumir independÃªncia entre variÃ¡veis. Como vemos no heat map acima, hÃ¡ vÃ¡rios pares de variÃ¡veis, cuja correlaÃ§Ã£o Ã© significativa. Por exemplo, PCC(slope,oldpeak)=-0.58, o que sugere um nÃ­vel significativo de correlaÃ§Ã£o (caso fossem independentes a correlaÃ§Ã£o seria zero). Por isso, estamos a ignorar relaÃ§Ãµes importantes entre variÃ¡veis, diminuindo a accuracy do modelo.

```python
for col in x.columns:
    sns.histplot(x[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

Outro problema Ã© assumir distribuiÃ§Ã£o Gaussiana de todas as variÃ¡veis. Como vemos acima, a variÃ¡vel age parece ter uma distribuiÃ§Ã£o similar Ã  Gaussiana, no entanto, em variÃ¡veis categÃ³ricas, como o sex, a distribuiÃ§Ã£o normal nÃ£o se adequa. A distribuiÃ§Ã£o Gaussiana Ã© mais adequada para variÃ¡veis contÃ­nuas, neste caso temos dados mistos (categÃ³ricos e contÃ­nuos).
