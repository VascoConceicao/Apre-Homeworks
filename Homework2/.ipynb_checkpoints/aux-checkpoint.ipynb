```python
# Imprime uma tabela

def printTable(table):
    for i in range(len(table)):
        for j in range(len(table[i])):
            print(f'{table[i][j]} ', end = '')
        print('\n', end = '')
```

```python
# Constrói o dataset

data  = []
data += [['P', 'A', 0]]
data += [['P', 'B', 1]]
data += [['P', 'A', 1]]
data += [['P', 'A', 0]]
data += [['N', 'B', 0]]
data += [['N', 'B', 0]]
data += [['N', 'A', 1]]
data += [['N', 'B', 1]]

printTable(data)
```

```python
# Calcula a distância de Hamming entre duas instâncias

def hammingDistance(x, y):
    distance = 0
    for i in range(1, len(x)):
        if (x[i] != y[i]):
            distance += 1
    return distance

def weightedHammingDistance(x, y):
    distance = 0
    weights = [0.10, 0.9]
    for i in range(1, len(x)):
        if (x[i] != y[i]):
            distance += weights[i - 1] * 1
    return distance

import math

def euclidianDistance(x, y): # mesma medida f1
    x1 = 0
    x2 = x[2]
    y1 = 0
    y2 = y[2]
    if x[1] == 'B':
        x1 = 1
    if y[1] == 'B':
        y1 = 1
    return math.sqrt((x1 - y1) ** 2 + (x2 - y2) ** 2)

def euclidianDistanceDummified(x, y): # TP = 0 (precision = 0 e recall = 0)
    x1 = 0
    x2 = 0
    x3 = x[2]
    y1 = 0
    y2 = 0
    y3 = y[2]
    if x[1] == 'A':
        x1 = 1
    elif x[1] == 'B':
        x2 = 1
    if y[1] == 'A':
        y1 = 1
    elif y[1] == 'B':
        y2 = 1
    return math.sqrt((x1 - y1) ** 2 + (x2 - y2) ** 2 + (x2 - y3) ** 2)

def manhattanDistance(x, y): # mesma medida f1
    x1 = 0
    x2 = x[2]
    y1 = 0
    y2 = y[2]
    if x[1] == 'B':
        x1 = 1
    if y[1] == 'B':
        y1 = 1
    return abs(x1 - y1) + abs(x2 - y2)

def manhattanDistanceDummified(x, y): # TP = 0 (precision = 0 e recall = 0)
    x1 = 0
    x2 = 0
    x3 = x[2]
    y1 = 0
    y2 = 0
    y3 = y[2]
    if x[1] == 'A':
        x1 = 1
    elif x[1] == 'B':
        x2 = 1
    if y[1] == 'A':
        y1 = 1
    elif y[1] == 'B':
        y2 = 1
    return abs(x1 - y1) + abs(x2 - y2) + abs(x2 - y3)

distance = weightedHammingDistance

# Constrói a tabela de distâncias do dataset

distanceTable = []
for i in range(len(data)):
    line = []
    for j in range(len(data)):
        line += [distance(data[i], data[j])]
    distanceTable += [line]

printTable(distanceTable)
```

```python
# Calcula a k-vizinhança da instância xi

def findKNN(k, i):
    count = []
    for j in range(1, len(data) + 1):
        if i != j:
            pair = [j, distance(data[i - 1], data[j - 1])]
            if len(count) < k:
                count += [pair]
            else:
                count = sorted(count, key=lambda x: x[1])
                if distance(data[i - 1], data[j - 1]) < count[-1][1]:
                    count[-1] = pair
    res = []
    for i in range(k):
        res += [count[i][0]]
    return sorted(res)

# Constrói uma tabela com a k-vizinhança de xi na linha i

k = 5
kNNTable = []
for i in range(1, len(data) + 1):
    kNNTable += [findKNN(k, i)]

printTable(kNNTable)
```

```python
# Faz uma previsão acerca da instância xi

def predictKNN(i):
    kNN = findKNN(k, i)
    p = 0
    n = 0
    for j in kNN:
        if data[j - 1][0] == 'P':
            p += 1
        elif data[j - 1][0] == 'N':
            n += 1
    if p > n:
        return 'P'
    elif p < n:
        return 'N'

# Constrói uma tabela com a previsão da instância xi na linha i

predictTable = []
for i in range(1, len(data) + 1):
    predictTable += [predictKNN(i)]

printTable(predictTable)
```

```python
# Constrói a matriz de confusão

def confusionMatrix():
    matrix = [[0 for _ in range(2)] for _ in range(2)]
    for i in range(len(data)):
        if data[i][0] == 'P' and predictKNN(i + 1) == 'P':
            matrix[0][0] += 1
        elif data[i][0] == 'P' and predictKNN(i + 1) == 'N':
            matrix[1][0] += 1
        elif data[i][0] == 'N' and predictKNN(i + 1) == 'P':
            matrix[0][1] += 1
        elif data[i][0] == 'N' and predictKNN(i + 1) == 'N':
            matrix[1][1] += 1

    return matrix

matrix = confusionMatrix()
printTable(matrix)
```

```python
for weight in range(0, 101):
    weights = [weight/100, (100-weight)/100]
    def weightedHammingDistance(x, y, weights):
        distance = 0
        for i in range(1, len(x)):
            if (x[i] != y[i]):
                distance += weights[i - 1] * 1
        return distance

    distance = weightedHammingDistance

    # Calcula a k-vizinhança da instância xi

    def findKNN(k, i):
        count = []
        for j in range(1, len(data) + 1):
            if i != j:
                pair = [j, distance(data[i - 1], data[j - 1], weights)]
                if len(count) < k:
                    count += [pair]
                else:
                    count = sorted(count, key=lambda x: x[1])
                    if distance(data[i - 1], data[j - 1], weights) < count[-1][1]:
                        count[-1] = pair
        res = []
        for i in range(k):
            res += [count[i][0]]
        return sorted(res)

    # Constrói uma tabela com a k-vizinhança de xi na linha i

    k = 5
    kNNTable = []
    for i in range(1, len(data) + 1):
        kNNTable += [findKNN(k, i)]
    
    # Constrói a matriz de confusão
    
    def confusionMatrix():
        matrix = [[0 for _ in range(2)] for _ in range(2)]
        for i in range(len(data)):
            if data[i][0] == 'P' and predictKNN(i + 1) == 'P':
                matrix[0][0] += 1
            elif data[i][0] == 'P' and predictKNN(i + 1) == 'N':
                matrix[1][0] += 1
            elif data[i][0] == 'N' and predictKNN(i + 1) == 'P':
                matrix[0][1] += 1
            elif data[i][0] == 'N' and predictKNN(i + 1) == 'N':
                matrix[1][1] += 1
    
        return matrix
    
    matrix = confusionMatrix()
    
    # Calcula a medida F1 e as métricas necessárias
    
    def recall():
        return matrix[0][0] / (matrix[0][0] + matrix[1][0])
    
    def precision():
        return matrix[0][0] / (matrix[0][0] + matrix[0][1])
    
    def f1Measure():
        return 1/(1/2 * (1/precision() + 1/recall()))
    
    print(weight, f1Measure())
```
